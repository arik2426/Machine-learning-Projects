# Iris data basic EDA

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
df
```

![](iris.jpg)

```python
attributes = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
label = ['species']
```

**Distributions**
<img src="quartiles.jpg" width="700">

```python
_,axs = plt.subplots(2,2,figsize = (15,8))
axs = axs.ravel() 
for n,attribute in enumerate(attributes):
    sns.histplot(df, kde=True,x=attribute, hue="species", bins=30, fill=True, ax = axs[n]);
```

```python
df.boxplot();
```

```python
df.boxplot(by='species',figsize=(10,6));
```

```python
sns.pairplot(df, vars=attributes);
```

```python
sns.pairplot(df, vars=attributes, hue="species");
```
