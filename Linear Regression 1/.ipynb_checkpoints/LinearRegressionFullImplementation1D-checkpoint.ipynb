```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = np.loadtxt('HousePriceVsSizeLinearRegression.txt',delimiter=',', skiprows = 1)
x = Data[:,0]
t = Data[:,1]
plt.plot(x,t,'.', color='red')
plt.title('House range')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show()
```

```python
x_mean = np.mean(x)
t_mean = np.mean(t)

numerator = np.sum((x - x_mean) * (t-t_mean))
demnominator = np.sum((x- x_mean) **2)
                   
w1 = numerator/ demnominator  
```

## W1

```python
w1
```

```python
w0 = t_mean - w1*x_mean
```

## W0

```python
w0
```

```python
plt.plot(x,t,'.', color='red')
plt.plot(x,(w0 + w1*x), color='green')
plt.title('House range')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show()
```

## Error

```python
error = ((t - (w0 +w1*x))**2).mean()
error
```

```python
split_index_1 = 35

x_train = x[:split_index_1]
t_train = t[:split_index_1]
x_test = x[split_index_1:]
t_test = t[split_index_1:]

x_train_mean = np.mean(x_train)
t_train_mean = np.mean(t_train)

numerator = np.sum((x_train - x_train_mean) * (t_train - t_train_mean))
denominator = np.sum((x_train - x_train_mean)**2)

w1_train = numerator / denominator  
w0_train = t_train_mean - w1_train * x_train_mean

train_error = ((t_train - (w0_train + w1_train * x_train))**2).mean()
test_error = ((t_test - (w0_train + w1_train * x_test))**2).mean()

plt.figure(figsize=(8, 6))
plt.plot(x_train, t_train, '.', label='Training Data', color='blue')
plt.plot(x_test, t_test, '.', label='Test Data', color='red')
plt.plot(x, w0_train + w1_train * x, color='green', label='Linear Regression')

plt.title(f'House range')
plt.xlabel('Size')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
split_indexes = [10, 3]

for split_index in split_indexes:
    x_train = x[:split_index]
    t_train = t[:split_index]
    x_test = x[split_index:]
    t_test = t[split_index:]

    x_train_mean = np.mean(x_train)
    t_train_mean = np.mean(t_train)

    numerator = np.sum((x_train - x_train_mean) * (t_train - t_train_mean))
    denominator = np.sum((x_train - x_train_mean)**2)

    w1_train = numerator / denominator  
    w0_train = t_train_mean - w1_train * x_train_mean

    train_error = ((t_train - (w0_train + w1_train * x_train))**2).mean()
    test_error = ((t_test - (w0_train + w1_train * x_test))**2).mean()

    plt.figure(figsize=(8, 6))
    plt.plot(x_train, t_train, '.', label='Training Data', color='blue')
    plt.plot(x_test, t_test, '.', label='Test Data', color='red')
    plt.plot(x, w0_train + w1_train * x, color='green', label='Linear Regression')

    plt.title(f'House range')
    plt.xlabel('Size')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
N = x.shape[0]
X = np.ones((N,2))
X[:,1] = x
X
```

```python
T = np.ones((N,1))
T[:,0] = t
T
```

## W*

```python
W = np.linalg.inv(X.T @ X) @ X.T @ T
W
```

## ERROR

```python
ERROR = ((T - X@W).T @ (T - X@W)) / N
ERROR
```
