```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig
```

```python
data = pd.read_csv('FacesPixels.csv' , header=None)
plt.scatter(data.iloc[:, 0], data.iloc[:, 1])
plt.title('Original Data')
plt.show()
```

```python
# Show the data
data
```

```python
mu = data.mean(axis=0)
data_centered = data - mu

plt.scatter(data_centered.iloc[:, 0], data_centered.iloc[:, 1])
plt.title('Centered Data')
plt.show()
```

```python
sigma = (data_centered.T)@data_centered
sigma
evals,evecs =np.linalg.eig(sigma)
```

```python
evals =evals.real
```

```python
evecs = evecs.real
```

```python
idx = evals.argsort()[::-1]
idx
```

```python
evals = evals[idx]
evals
```

```python
plt.semilogy(idx,evals)
```
