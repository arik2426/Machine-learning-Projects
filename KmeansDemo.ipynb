```python
def OurKmeans(X,K):
    cluster1 =np.zeros(X.shape[0])
    cluster =np.ones(X.shape[0])

    ind = np.random.permutation(150)[:K]
    C = X[ind]

    while (cluster1==cluster).sum()<len(cluster):
        d = spatial.distance.cdist(X,C)
        cluster1 = cluster
        cluster = d.argmin(axis=1)
        
        for k in range(K):
            C[k] = X[np.where(cluster==k)[0]].mean(axis=0)
    
    return cluster
```
